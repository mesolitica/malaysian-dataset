```python
import os

os.environ['CUDA_VISIBLE_DEVICES'] = '0'
os.environ['TF_FORCE_GPU_ALLOW_GROWTH'] = 'true'
```

```python
from glob import glob
```

```python
files = sorted(glob('salina/*/*.wav'))
len(files)
```

```python
import IPython.display as ipd
ipd.Audio(files[0])
```

```python
import malaya_speech
```

```python
model = malaya_speech.stt.transducer.pt_transformer(model = 'mesolitica/conformer-medium')
```

```python
_ = model.cuda()
```

```python
y, _ = malaya_speech.load(files[0])
```

```python
model.beam_decoder([y])
```

```python
from tqdm import tqdm

results = {}
for f in tqdm(files):
    y, _ = malaya_speech.load(f)
    r = model.beam_decoder([y])[0]
    results[f] = r
```

```python
import json

with open('semisupervised-salina.json', 'w') as fopen:
    json.dump(results, fopen)
```
