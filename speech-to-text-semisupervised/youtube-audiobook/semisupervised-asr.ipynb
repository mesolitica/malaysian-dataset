```python
import malaya_speech
```

```python
model = malaya_speech.stt.transducer.pt_transformer('mesolitica/conformer-medium')
```

```python
_ = model.cuda()
```

```python
from datasets import Audio

audio = Audio(sampling_rate=16000)
```

```python
from glob import glob
from tqdm import tqdm

files = sorted(glob('*/*.mp3'))
len(files)
```

```python
import IPython.display as ipd
ipd.Audio('(Audiobook) Mengenali Imam Mahdi dan Akhir Zaman - Bahagian 1 [XpSalODXHKA]_/102.mp3')
```

```python
results = {}

for f in tqdm(files):
    r = audio.decode_example(audio.encode_example(f))['array']
    o = model([r])[0][0]
    results[f] = o
```

```python
import json

with open('semisupervised-asr.json', 'w') as fopen:
    json.dump(results, fopen)
```
