```python
with open('a.json') as fopen:
    nested_a = json.load(fopen)
nested_a = sorted(list(set(nested_a)))
```

```python
for no, url in tqdm(enumerate(nested_a)):
    filename = os.path.join('page', f'{no}.json')
    if os.path.exists(filename):
        continue
    
    while True:
        try:
            r = requests.get(url)
            break
        except:
            time.sleep(1.0)
            
    soup = BeautifulSoup(r.content, "lxml")
    
    try:
        title = soup.find('a', {'id':"titlehref"}).text
    except:
        title = None
    
    try:
        div = soup.find('div', {'style': "padding:20px; padding-left:10px;font-family:monospace;font-size:10pt;"})
        body = div.get_text(separator="\n")
    except:
        body = None

    data = {
        'url': url,
        'title': title,
        'body': body,
    }
    with open(filename, 'w') as fopen:
        json.dump(data, fopen)
```
