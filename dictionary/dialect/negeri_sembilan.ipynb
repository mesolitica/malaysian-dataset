```python
import json
import pandas as pd
from bs4 import BeautifulSoup
```

```python
with open('negeri_sembilan.json') as fopen:
    n9 = json.load(fopen)
```

```python
res = []

for j in n9:
    s = BeautifulSoup(j)
    s.tbody.table.decompose()
    table_rows = s.tbody.find_all('tr')
    for tr in table_rows:
        td = tr.find_all('td')
        row = [tr.text.strip() for tr in td if tr.text.strip()]
        if row:
            res.append(row)
```

```python
df = pd.DataFrame(res)
df
```

```python
df.to_csv('negeri_sembilan.csv', index = False)
```
